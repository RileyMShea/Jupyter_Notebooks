```python
#to open urls
from urllib import urlopen
#beautiful soup
from bs4 import BeautifulSoup
```

```python
url = "http://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
```

```python
#Create a soup object from the webpage
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
# Print out the text
text = soup.get_text()
print(soup.text)
```

```python
#return all links in the webpage 
soup.find_all('a')
```

```python
#use href to only print out hyperlinks
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
# Print the first 10 rows for sanity check
rows = soup.find_all('tr')
print(rows[:10])
```

```python
cleantext = []
for row in rows:
    row_td = row.find_all('td')
    str_cells = str(row_td)
    cleantext.append(BeautifulSoup(str_cells, "lxml").get_text())
print(cleantext)
type(cleantext)
```

```python
#extract the text without html tags
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
cleantext = str(cleantext)
print(cleantext)
type(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```
